- Concatenate numerical and categorical back together again for your X dataframe.  Designate the TargetB as y.

```python
donors = donors.drop("MDMAUD", axis=1)
donors 
```

```python
y = donors["TARGET_B"]
X = donors.drop(['TARGET_B'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 1)
```

 - Split further into train_num and train_cat.  Also test_num and test_cat.

```python
train_num = X_train.select_dtypes(include=[np.number])
```

```python
train_cat = X_train.select_dtypes(include='object')
train_cat.shape
```

```python
test_num =  X_test.select_dtypes(include=[np.number])
```

```python
test_cat = X_test.select_dtypes(include='object')
test_cat.shape
```

- Scale the features either by using MinMax Scaler or a Standard Scaler. (train_num, test_num)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
transformer = MinMaxScaler().fit(train_num)
```

```python
train_num_norm = pd.DataFrame(transformer.transform(train_num), columns=train_num.columns)
test_num_norm = pd.DataFrame(transformer.transform(test_num), columns=test_num.columns)
```

REDUCING THE NUMERICAL FEATURES

```python
from sklearn.feature_selection import VarianceThreshold 
var_threshold = 0.1 #PERSONAL DECISSION (WHY? WHY NOT)
sel = VarianceThreshold(threshold=(var_threshold))
```

```python
sel = sel.fit(train_num_norm)
temp = sel.transform(train_num_norm)
selected_train_num = pd.DataFrame(temp)
print(train_num_norm.shape)
print(selected_train_num.shape)
```

```python
selected_test_num = pd.DataFrame(sel.transform(test_num_norm))
print(test_num_norm.shape)
print(selected_test_num.shape)
```

  - Encode the categorical features using One-Hot Encoding or Ordinal Encoding.  (train_cat, test_cat)
      - **fit** only on train data, transform both train and test

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder().fit(train_cat) 
```

```python
unique_values = [encoder.categories_[i] for i in range(len(train_cat.columns))]
train_encoded = pd.DataFrame(encoder.transform(train_cat).toarray(),columns=[f"{col}_{value}" for col, values in zip(train_cat.columns, unique_values) for value in values])
test_encoded = pd.DataFrame(encoder.transform(test_cat).toarray(),columns=[f"{col}_{value}" for col, values in zip(train_cat.columns, unique_values) for value in values])
```

 - again re-concatenate train_num and train_cat as X_train as well as test_num and test_cat as X_test

```python
X_train = pd.concat([selected_train_num, train_encoded], axis=1)
X_train
```

```python
X_test = pd.concat([selected_test_num, test_encoded], axis=1)
X_test
```

  - Fit a logistic regression model on the training data.

```python
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)
```

```python
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

  - Check the accuracy on the test data.

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
```

```python
acc = accuracy_score(y_test, y_pred)
acc
```

```python
recall = recall_score(y_test, y_pred)
recall
```

```python
f1 = f1_score(y_test, y_pred)
f1
```

```python
classification_report = classification_report(y_test, y_pred)
classification_report 
```

# Managing imbalance in the dataset

- Check for the imbalance.

```python
y_train = y_train.reset_index(drop=True)
```

```python
train = pd.concat([X_train, y_train], axis=1)
train
```

```python
train['TARGET_B'].value_counts()
```

```python
from sklearn.utils import resample

category_0 = train[train['TARGET_B'] == 0]
category_1 = train[train['TARGET_B'] == 1]
```

UNDERSAMPLE

```python
category_0_undersampled = resample(category_0,
                                   replace=False,
                                   n_samples = len(category_1))
```

```python
print(category_0_undersampled.shape)
print(category_1.shape)
```

```python
data_downsampled = pd.concat([category_0_undersampled, category_1], axis=0)
```

OVERSAMPLE

```python
category_1_oversampled = resample(category_1,
                                  replace=True,
                                  n_samples = len(category_0))
```

```python
print(category_0.shape)
print(category_1_oversampled.shape)
```

```python
data_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)
```

- Each time fit the model and see how the accuracy of the model has changed.

LOGISTIC REGRESSION:  UNDER SAMPLE

```python
X_train_down = data_downsampled.drop("TARGET_B", axis=1)
y_train_down = data_downsampled["TARGET_B"]
```

```python
classifier_down = LogisticRegression()
classifier_down.fit(X_train_down, y_train_down)
```

```python
y_pred_down = classifier_down.predict(X_test)
```

```python
acc_down = accuracy_score(y_test, y_pred_down)
acc_down
```

```python
recall_down = recall_score(y_test, y_pred_down)
recall_down
```

```python
f1_down = f1_score(y_test, y_pred_down)
f1_down
```

LOGISTIC REGRESSION:  OVER SAMPLE

```python
X_train_up = data_upsampled.drop("TARGET_B", axis=1)
y_train_up = data_upsampled["TARGET_B"]
```

```python
classifier_up = LogisticRegression()
classifier_up.fit(X_train_up, y_train_up)
```

```python
y_pred_up = classifier_up.predict(X_test)
```

```python
acc_up = accuracy_score(y_test, y_pred_up)
acc_up
```
